**Task 1. Import all the libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#the line above is to run matplotlib in this notebook
```

**Task 2: Loading Dataset**

```python
df = pd.read_csv("datasets.csv", encoding_errors = "ignore") #read the csv file into the DataFrame
```

**Task 4: Dara Cleaning**

```python
df.dropna(inplace=True) #dropping all missing values rows
df.isnull().sum() #is used to remove missing values

#df.fillna() is used to fill missing values
```

```python
df.shape # gives the dimension of DataFrame(no. of rows , no. of columns)
```

```python
df.duplicated().sum() #dealing with duplicates
df.drop_duplicates(inplace=True)#deleating all duplicates
```

```python
df.dtypes # Shows the datatype of each column
```

```python
#convert the datatype of one or more column
df = df.astype({
    'id':'object',
    'host_id':'object',
    'beds':'object',
    })

df.dtypes  # Shows the datatype of each column
```

## EDA
**Task 5: Data Analysis**


**Univariate Analysis**

```python
# identifying outliers in price
df = df[df['price']<1500]
sns.boxplot(data=df,x='price')
```

```python
# Analysing price distribution
plt.figure(figsize =(8,5)) #create a new figure with specified size
sns.histplot(data=df, x ='price',bins =100) #plot a histogram of price column
plt.title("Price Distribution") #add a title to the plot
plt.ylabel("Frequency") #label the Y-axis
plt.show() # Display the plot
```

```python
df.columns # Display the column names of the dataframe
```

```python
df.dtypes # check the data types of each column
```

```python
# Analysing price distribution
plt.figure(figsize =(6,3)) #create a new figure with specified size
sns.histplot(data=df, x ='availability_365') #plot a histogram
plt.title("availability_365 Distribution") #add a title to the plot
plt.ylabel("Frequency") #label the Y-axis
plt.show() # Display the plot
```

```python
df.groupby(by = 'neighbourhood_group')[['price','number_of_reviews']].mean() #grouped rows based on one or more columns here in terms of price and number of reviews
```

```python
df['price_per_bed'] = df['price']/df['beds'] # calculated 'price_per_bed'
#print(df['price_per_bed'])
df.head()
```

```python
df.groupby(by = 'neighbourhood_group')['price_per_bed'].mean() #avg price per bed
```

```python
df.groupby(by = 'neighbourhood_group')['price'].mean() #avg price per bed
```

**Bi variable Analysis**
One variable dependency in another variable



```python
sns.barplot(data=df, x='neighbourhood_group',y='price', hue = 'room_type') #price dependency on neighbourhood_group
```

```python
# number of reviews and price
plt.figure(figsize = (8,5))  #create a new figure with specified size
plt.title("Reviews and price data based on beds for Manhattan")  #add a title to the plot
sns.scatterplot(data=df[df['neighbourhood_group']=='Manhattan'],x ='price' ,y='number_of_reviews',hue = 'beds') #created a scatterplot to visualize the relationship between price and number of reviews
plt.show()
```

```python
df.dtypes
```

```python
sns.pairplot(data=df,vars=['price','minimum_nights','number_of_reviews','availability_365'],hue = 'room_type') #creating a grid of scatter plots and histograms for visualizing pairwise relationship between price,minimum_nights,number_of_reviews,availability_365.
```

```python
#geographical distribution of Airbnb
plt.figure(figsize = (10,8))
plt.title("Geographical Distribution")
sns.scatterplot(data=df, x = 'longitude', y ='latitude', hue = 'room_type')
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.show()
```

```python
df.dtypes
```

```python
#heat map - correlation of one varialble with others for numerical column

corr =df[['latitude','longitude','price','minimum_nights','reviews_per_month','availability_365','beds']].corr()
corr

plt.figure(figsize =(8,8))
plt.title('Correlation Heatmap')
sns.heatmap(data =corr , annot =True)
plt.show()
```
